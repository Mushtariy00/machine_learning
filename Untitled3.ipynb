```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
path = '/Users/mushtariy/pyprojects/Machine_Learning/machine_learning/NFLX.csv'
```

```python
df = pd.read_csv(path)
df.head(10)
```

```python
## Adding a new column with reindexing
df.reset_index(inplace=True)
df.head(10)
```

```python
# making a Correlation columnt
corr = df.corr(method='pearson')
corr
```

```python
# making a heatmap of dataset where dark denotes the highly correlated features
import seaborn as sb
sb.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns,
          cmap='RdBu_r', annot=True, linewidth=0.5)
```

```python
nflx_df = df[['Date', 'High', 'Open', 'Low', 'Close']]
nflx_df.head(10)
```

```python
nflx_df[['Open','Close']].head(20).plot(kind='bar',figsize=(16,8))
plt.show()
```

```python
df['Date'] = pd.to_datetime(df.Date, format='%Y-%m-%d')
```

```python
nflx_df['Year']=df['Date'].dt.year
nflx_df['Month']=df['Date'].dt.month
nflx_df['Day']=df['Date'].dt.day
```

```python
nfx_df=nflx_df[['Day', 'Month', 'Year', 'High', 'Open', 'Low', 'Close']]
nfx_df.head(10)
```

```python
X = nfx_df.iloc[:,nfx_df.columns !='Close']
Y= nfx_df.iloc[:, 6]
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y,random_state=0, test_size=.25)
```

```python
lr_model = LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
y_pred = lr_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
plot_df=pd.DataFrame({'Actual':y_test, 'Pred':y_pred})
plot_df.head(20).plot(kind = 'bar', figsize=(16,8))
plt.show()
```
