## Reading a dataset based on Pandas

```python
import pandas as pd
df_iris = pd.read_csv('dataset/Iris.csv')
```

- PetalLengthCm
- PetalWidthCm

## Calculating missing values on the dataset

#### Checking if there is a NaN values.

```python
df_iris.isnull().values.any()
```

```python
missing_values = df_iris.isnull().sum().sum()
```

```python
print(f'Total missing values are: {missing_values}')
```

## Replacing missing values

#### Replacing NaN values for single column

```python
df_iris['SepalLengthCm'] = df_iris['SepalLengthCm'].fillna(0)
```

#### Checking if there is a NaN values in a 'SepalLengthCm' column.

```python
df_iris['SepalLengthCm'].isnull().values.any()
```

#### Checking the number of NaN values

```python
df_iris['SepalLengthCm'].isnull().sum().sum()
```

#### Replacing all NaN values with zeros(0)

```python
df_iris = df_iris.fillna(0)
```

#### Checking if there is a NaN values.

```python
df_iris.isnull().values.any()
```

```python
print(df_iris.to_string())
```

## Preparing a bar chart

```python
df_iris.plot(kind = 'bar', y = ["SepalLengthCm","SepalWidthCm", "PetalLengthCm", 
              "PetalWidthCm", "Species"], title='Iris Dataset')
```

```python
df_iris.plot(kind = 'area', y = ["SepalLengthCm","SepalWidthCm", "PetalLengthCm", 
              "PetalWidthCm", "Species"], title='Iris Dataset')
```
